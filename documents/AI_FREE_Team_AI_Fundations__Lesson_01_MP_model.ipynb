# 類神經網路的始祖

類神經網路的起源，可追朔至西元1943年，由 Warren McCulloch 和 Walter Pitts 所推出的McCulloch–Pitts neuron network，簡稱 MP model；在當時電腦科技起飛，科學家們不斷苦思人工智慧存在的可能性，因此透過數學模型模擬了神經細胞傳導的模式，也因此被稱作類神經網路。

# 模型邏輯

簡單來說，MP model 模擬神經元接收外部資訊(參數x - 0 或 1)，透過數學加總與邏輯判斷後，

最後在模擬神經元傳遞處理的資訊 - 輸出判斷結果(參數y -  0 或 1)，全部參數 x & y 都是 boolen 值。

# 邏輯運算

在 MP model 中，主要講述三個邏輯運算 AND (&) 、 OR (|) 、 XOR (^)



*   AND (&)：輸入參數(x)皆為 1 → 觸發模型參數(y)為 1 (否則為0)
*   OR (|)：輸入參數(x)其中含 1 → 觸發模型參數(y)為 1 (否則為0)
*   XOR (^)：輸入參數(x)不具 1 → 觸發模型參數(y)為 1 (否則為0)


# 程式碼

知道了以上三個邏輯運算的概念後，我們將透過實驗，來將邏輯判斷的臨界值 (threshold) 找出。

## 建立簡易 4(row) * 2(column) 的矩陣。

(共四筆不同的 x 參數數據)

```
import numpy as np
x = np.array([[0,0],[0,1],[1,0],[1,1]])
print(x)
```

## 建立邏輯運算出來的解， 4(row) * 1(column)的矩陣。


```
y_and = np.array([0,0,0,1])
print('y_and: ', y_and)

y_or = np.array([0,1,1,1])
print('y_or: ', y_or)

y_xor = np.array([0,1,1,0])
print('y_xor: ', y_xor)
```

## 建構我們 MP model 的實驗

```
class MP_Neuron_network:

  # 初始化
  def __init__(self):
    self.b = None

  # 數學判斷
  def math_f(self, x):
    return(sum(x) >= self.b) # 回傳 x 的加總是否大於b

  # 將數學判斷結果記錄到 Y
  def predict(self, X):
    Y = []
    for x in X:
      result = self.math_f(x)
      Y.append(result)
    return np.array(Y)

  # 計算判斷的成功率
  def acc_score(self, y_predict, y_real):
    c = 0
    for i,j in zip(y_predict, y_real):
      if i == j:
        c += 1
    return (c/len(y_real))*100

  # MP model 實驗開始
  def fit(self, X, Y):
    # 建立空的字典，作為後續紀錄實驗評分
    accuracy = {}

    # 以迴圈定義 b 值 (0 到 4)
    for b in range(5):
      self.b = b

      # 將邏輯判斷結果記錄至 Y_pred
      Y_pred = self.predict(X)

      # 將結果做評分並記錄至 accuracy
      accuracy[b] = self.acc_score(Y_pred, Y)
      print('當 b 為 {} ，預測結果為 {} ，判斷正確率為 {}'.format(b, Y_pred, accuracy[b]))

      best_b = max(accuracy, key = accuracy.get)
      print('(--- 目前最佳解：當 b 為 {}，正確率為 {} ---)'.format(best_b, accuracy[best_b]))
      print("")
      self.b = best_b

    print('當 b 為 {} ，實驗判斷結果最佳！'.format(best_b))
    print('正確率為： ',accuracy[best_b])
```

## 實驗驗證 (找到判斷的臨界點)

### AND (&)

```
mp_model = MP_Neuron_network()
mp_model.fit(x, y_and)
```

### OR (|)

```
mp_model = MP_Neuron_network()
mp_model.fit(x, y_or)
```

### XOR (^)

```
mp_model = MP_Neuron_network()
mp_model.fit(x, y_xor)
```

#結論與發現

MP model 在 XOR 的運算判斷無法找到正確率 100% 的 b 值，是因為當函數(邏輯判斷)為非線性時，

MP model 無法有效地進行判斷，因此我們需要更複雜的類神經模型，而不僅是加總後透過臨界值判斷。
